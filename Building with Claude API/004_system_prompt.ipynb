```python
#Load Env Variables

from dotenv import load_dotenv

load_dotenv()


#Anthropic SDK
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
def add_user_message(messages, text):
    user_message = {"role":"user", "content" : text}
    messages.append(user_message)
    
def add_assistant_message(messages, text):
    assistant_message = {"role":"assistant", "content" : text}
    messages.append(assistant_message)
    
def chat(messages,system=None, temperature = 1.0):
    params = {
        "model" : model,
        "max_tokens" : 1000,
        "messages" : messages,
        "temperature" : temperature
    }

    if system:
        params["system"] = system
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
messages = []

# system_prompt = """
#     You are a patient and helpful math tutor.
#     Do not directly answer a student's questions.
#     Guide them to a solution step by step, with details
#     """

add_user_message(messages, "How do i solve 5x+3=2 for x?")
answer = chat(messages)

print(answer)
```

```python
##System Prompt Exercise

messages = []
system_prompt = """
    You are a helpful python coding assistant , 
    when ever user asks for python function or python implementation
    give them the direct code without unneccessary details
"""
add_user_message(
    messages,
    "write a Python function that checks a string for duplicate characters.",
)

answer = chat(messages, system = system_prompt)

print(answer)
```
